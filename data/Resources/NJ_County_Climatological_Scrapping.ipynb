```python
#dependencies
import pandas as pd
from bs4 import BeautifulSoup as bs
import time
from splinter import Browser
import csv
```

```python
executable_path = {'executable_path': '/usr/local/bin/chromedriver'}
browser = Browser('chrome', **executable_path, headless=True)
```

```python
#Atlantic County
url= 'https://climate.rutgers.edu/stateclim_v1/nclimdiv/index.php?stn=NJ001&elem=avgt'

browser.visit(url)
time.sleep(2)

html= browser.html
soup= bs(html, 'html.parser')

table= soup.find('table', id='climdiv_table')

# Annual data
weather_df = pd.read_html(str(table))
weather_df= weather_df[0]

avg_yrs = weather_df[["Year", "Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]]
avg_yrs = avg_yrs.iloc[75:125,]

avg_yrs.insert(1, "County", "Atlantic")

avg_yrs.to_csv('Atlantic County Monthly Average Temperature Across Years.csv')


# Mean accross the months 
mean_df= weather_df.iloc[128:129,]
mean_df= mean_df.drop(columns=['Rank', 'Year','Annual'])
mean_df.insert(0, "County", "Atlantic")
mean_df.to_csv('Atlantic County Mean Monthly Temperature.csv')
```
