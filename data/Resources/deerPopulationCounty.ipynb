# Deer Population by County

```python
#Import dependencies
import os
import pandas as pd
```

```python
#File paths
deer_population_pdf_path = os.path.join("","Resources","rawData","NJFB-SG-State-Report_10.2019_1.pdf")
summary_data =  os.path.join("","Resources","summaryData","tabula-NJFB-SG-State-Report_10.2019_1.csv")
```

### PDF data extractu- tabula library

```python
#Import dependencies
import tabula
```

```python
#Read pdf file
deer_pop_df = tabula.read_pdf(deer_population_pdf_path,output_format='dataframe', pages = "all", multiple_tables = True)
```

```python
# Export tables to a CSV
tabula.convert_into(deer_population_pdf_path, summary_data, pages = "all")
```

### PDF data PyPDF2 library

```python
# pip install PyPDF2
```

```python
#Import dependencies
import PyPDF2
```

```python
# creating a pdf file object 
pdfFileObj = open(deer_population_pdf_path, 'rb')
```

```python
# creating a pdf reader object 
pdfReader = PyPDF2.PdfFileReader(pdfFileObj) 
```

```python
# printing number of pages in pdf file 
print(pdfReader.numPages) 
```

```python
# creating a page object 
pageObj = pdfReader.getPage(9) 
```

```python
# extracting text from page 
print(pageObj.extractText()) 
```

```python
# closing the pdf file object 
pdfFileObj.close() 
```
