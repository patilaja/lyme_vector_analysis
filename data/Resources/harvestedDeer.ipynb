# Summary of Deer Harvested by Season and Year 1972-73 Through 2019-20  

```python
#Import Dependencies
import pandas as pd
import os
```

```python
#File paths
tables = pd.read_html("https://www.state.nj.us/dep/fgw/deerharvestsummary.htm")
summary_data =  os.path.join("","Resources","summaryData","summaryDeerHarvested_Season_Year.csv")
```

```python
#Create data frame from the html content
deerHarvestedSeason=pd.DataFrame(tables[6])
```

```python
#Sample data
deerHarvestedSeason.head(5)
```

```python
#Get column names
deerHarvestedSeason.columns
```

```python
#Rename the data frame columns for individual quarter data
deerHarvestedSeason= deerHarvestedSeason.rename(columns={0: "Season", 1: "Fall Bow",
                   2:"Youth Hunts",3:"Permit Bow",4:"Six-day Firearm",5:"Permit Muzzleloader",
                    6:"Permit Shotgun",7:"Winter Bow",8:"Total Harvest",9:"Antlered Harvest",
                    10:"Antlerless Harvest"})
```

```python
#Display sample results
deerHarvestedSeason.head(3)
```

```python
#Filter duplicated header rows from the data
deerHarvestedSeason = deerHarvestedSeason[deerHarvestedSeason['Season'] != 'Seasons'] 
```

```python
#Fill NA values with 0
deerHarvestedSeason.fillna(0)
```

```python
#Export data to csv file
deerHarvestedSeason.to_csv(summary_data,index=False)
```
